```python
data = pd.read_csv("SHORE_SITE_LIST_NTCSS-OOMA_For_Scripts.csv")
low_memory = False
```

```python
df = pd.DataFrame(data)
df
```

```python
def write_homeport():
    # Add Shore Sites and Ships to Refined Homeport variable file
    # add quotes around each element in the list
    df.update(df[['Refined Command Name', 'Refined Homeport']].applymap('"{}"'.format))

    f = open("C:/Users/barnesll/Desktop/Dimension - Refined Homeport_test.txt", "w")
    f.write("\n")
    i = 0
    while i < len(df):
        f.write("  ElseIf [Refined Command Name] = {0}".format(df['Refined Command Name'].iloc[i]) + " " + "Then" + " " + df['Refined Homeport'].iloc[i])
        f.write("\n")
        i = i + 1
    f.close()
```

```python
def write_commandjs():
    # Write Shore Sites and Ships to Command_Selection.js
    df.update(df[['Underscore']].applymap('"{}"'.format))

    f = open("C:/Users/barnesll/Desktop/Command_Selection_test.js", "w")
    f.write("\n")
    # <option value="MEF_CAMP_PENDLETON">MEF Camp Pendleton</option>
    j = 0
    while j < len(df):
        f.write("    <option value={0}".format(df['Underscore'].iloc[j]) + ">" + df['Proper Case'].iloc[j] + "</option>" )
        f.write("\n")
        j = j + 1
    f.close()
```

```python
def write_commandName():
    #Write Shore Sites to Refined Command Name variable file
    df.update(df[['Uppercase']].applymap('"{}"'.format))

    f = open("C:/Users/barnesll/Desktop/Dimension - Refined Command Name_test.txt", "w")
    f.write("\n")
    u = 0
    while u < len(df):
        if df['Refined Command Name'].iloc[u] == df['Uppercase'].iloc[u]:
            pass
        else:
            f.write("    {0}".format(df['Uppercase'].iloc[u]) + ";" + " " + df['Refined Command Name'].iloc[u] + ");") 
            f.write("\n")
        u = u + 1
    f.close()
```

```python
def write_Replace():
    # Write "(Replace" repetitively 
    f = open("C:/Users/barnesll/Desktop/replace.txt", "w")
    p = 1
    while p < 53:
        f.write("(Replace")
        p = p + 1
    f.close()
```
