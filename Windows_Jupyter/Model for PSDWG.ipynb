```python
%%time
# Calculate Median and Mean for all CB-ISEAs
cbisea_array = ['APPLICATIONS', 'CYBER SECURITY', 'ISR/IO', 'MARITIME/EXPEDITIONARY C2', 'MESSAGING', 'METOC', 'NAVIGATION', 'NC3',
               'RADIO', 'SATCOM', 'SHORE NET', 'SUB RADIO ROOMS', 'TACNET', 'TACTICAL DATA LINKS']
i = 0
incident_array = []
while i < len(cbisea_array):
    j = 0
    while j < len(df7):
        if df7['CB-ISEA'].iloc[j] == cbisea_array[i]:
            incident_array.append(df7['Incident Age (Days)'].iloc[j])
        else:
            pass
        j = j + 1
     
    print("CB-ISEA: " + cbisea_array[i])
    #print("Incident Ages: " + str(incident_array))
    print("Median: " + str(statistics.median(map(float, incident_array))))
    print("Mean: " + str(statistics.mean(map(float, incident_array))))
    
    #frequency = collections.Counter(incident_array)
    #frequency_sorted_keys = sorted(frequency, key=frequency.get, reverse=True)
    #for r in frequency_sorted_keys:
        #print(r, frequency[r])
    print('*******************************************')
    i = i + 1
    del incident_array[:]
```
