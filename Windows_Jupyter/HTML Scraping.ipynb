```python
soup = BeautifulSoup(html, "html.parser")
type(soup)
```

```python
title = soup.title
print(title)
```

```python
#Print out the text
text = soup.get_text()
print(soup.text)
```

```python
# Extract useful html tags
soup.find_all('sh_0')
```

```python
# Print out only hyperlinks
all_links = soup.find_all("fa_0")
for link in all_links:
    print(link.get("href"))
```

```python
# Print out table rows only
rows = soup.find_all('tr')
print(rows[:10])
```

```python
# Iterate through table rows and print out the cell of the rows
for row in rows:
    row_td = row.find_all('td')
print(row_td)
type(row_td)
```

```python
# Remove html tags
str_cells = str(row_td)
cleantext = BeautifulSoup(str_cells, "html.parser").get_text()
print(cleantext)
```

```python
import re

list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells))
    list_rows.append(clean2)
print(clean2)
type(clean2)
```

```python
# Convert list into dataframe
df = pd.DataFrame(list_rows)
df.head(10)
```

```python
# Cleanup the dataframe
df1 = df[0].str.split(',', expand=True)
df1.head(10)
```

```python
df1[0] = df1[0].str.strip('[')
df1.head(10)
```

```python
# Get table headers
col_labels = soup.find_all('th')
```

```python
all_header = []
col_str = str(col_labels)
cleantext2 = BeautifulSoup(col_str, "html.parser").get_text()
all_header.append(cleantext2)
print(all_header)
```

```python
# Convert list of headers into a dataframe
df2 = pd.DataFrame(all_header)
df2.head()
```

```python
df3 = df2[0].str.split(',', expand=True)
df3.head()
```

```python
frames = [df3, df1]

df4 = pd.concat(frames)
df4.head(10)
```

```python
df5 = df4.rename(columns=df4.iloc[0])
df5.head()
```
